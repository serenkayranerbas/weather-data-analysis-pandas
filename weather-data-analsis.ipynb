```python
#3.Sütun Seçimi:
#Sadece Date, City ve Temperature sütunlarını seçerek yeni bir görünüm oluşturun ve ekrana yazdırın.
#City ve Temperature kolonlarını birlikte listeleyiniz.
```

```python
#4.Basit Filtreleme:
#Hava sıcaklığı (Temperature) 30 derecenin üzerinde olan kayıtları listeleyin.
#Şehri (City) sadece "Bursa" olan kayıtları listeleyin.
```

```python
df[df["Temperature"] > 30] #Hava sıcaklığı (Temperature) 30 derecenin üzerinde olan kayıtları listeleyin.
```

```python
df[df['City']=='Bursa'] #Şehri (City) sadece "Bursa" olan kayıtları listeleyin.
```

```python
#5.Mantıksal Operatörler ile Filtreleme:
#Şehri "İstanbul" olan VE Nem oranı (Humidity) 60'tan büyük olan kayıtları listeleyin.
#Şehri "Ankara" olan VEYA Sıcaklığı 5 dereceden küçük olan kayıtları listeleyin.
#Sıcaklığı 10°C altında veya nemi (Humidity) %70 üzerinde olan verileri filtreleyiniz.
```

```python
df[(df['City']=='İstanbul') & (df['Humidity']>60)]  #Şehri "İstanbul" olan VE Nem oranı (Humidity) 60'tan büyük olan kayıtları listeleyin.
```

```python
df[(df["City"] == "Ankara") | (df["Temperature"] < 5)] #Şehri "Ankara" olan VEYA Sıcaklığı 5 dereceden küçük olan kayıtları listeleyin.
```

```python
df[(df["Temperature"] < 10) | (df["Humidity"] > 70)] #Sıcaklığı 10°C altında veya nemi (Humidity) %70 üzerinde olan verileri filtreleyiniz.
```

```python
#6.Sıralama (Sorting):
#Veri setini Temperature (Sıcaklık) değerine göre büyükten küçüğe (azalan) sıralayın ve ilk 10 kaydı gösterin.
#Veriyi neme göre azalan şekilde sıralayınız.
#Veriyi Şehir alanına göre Artan şekilde sıralayanız.
```

```python
df.sort_values(by="Temperature", ascending=False).head(10) #Veri setini Temperature (Sıcaklık) değerine göre büyükten küçüğe (azalan) sıralayın ve ilk 10 kaydı gösterin.
```

```python
df.sort_values(by="Humidity", ascending=False) #Veriyi neme göre azalan şekilde sıralayınız.
```

```python
df.sort_values(by="City", ascending=True) #Veriyi Şehir alanına göre Artan şekilde sıralayanız.
```

```python
#7.Yeni Sütun Ekleme:
#Veri setine Temperature_F adında yeni bir sütun ekleyin. Bu sütun, sıcaklığın Fahrenheit cinsinden değerini taşımalıdır.
#Formül: (Temperature * 9/5) + 32
#"FeelsLike" isminde yeni bir kolon oluşturun. Formül:
#FeelsLike = Temperature - (Humidity / 100)
```

```python
df["Temperature_F"] = (df["Temperature"] * 9/5) + 32  #Veri setine Temperature_F adında yeni bir sütun ekleyin. Bu sütun, sıcaklığın Fahrenheit cinsinden değerini taşımalıdır.
```

```python
df["FeelsLike"] = df["Temperature"] - (df["Humidity"] / 100) #"FeelsLike" isminde yeni bir kolon oluşturun. Formül:
```

```python
#8.Gruplama (Grouping) ve Analiz:
#Her şehirde (City) kaç adet veri kaydı olduğunu hesaplayın.
#Şehirlere göre ortalama sıcaklık (Temperature) değerlerini hesaplayın.
```

```python
df.groupby("City").size() #Her şehirde (City) kaç adet veri kaydı olduğunu hesaplayın.
```

```python
df.groupby("City")["Temperature"].mean() #Şehirlere göre ortalama sıcaklık (Temperature) değerlerini hesaplayın.
```

```python
#9.En Yüksek/Düşük Değer Analizi:
#En yüksek sıcaklığın olduğu satırı (tüm bilgilerini) bulun ve yazdırın.
#En düşük nem oranının olduğu satırı bulun ve yazdırın.
```

```python
df.loc[df["Temperature"].idxmax()] #En yüksek sıcaklığın olduğu satırı (tüm bilgilerini) bulun ve yazdırın.
```

```python
df.loc[df["Humidity"].idxmin()] #En düşük nem oranının olduğu satırı bulun ve yazdırın.
```
